### Feature Scaling
#### 1. Standardization

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
import numpy as np
```

```python
mean = np.mean(df['total_bill'])
std = np.std(df['total_bill'])
```

```python
mean,std
```

```python
normalized_data = []
for i in list(df['total_bill']):
    z_score = (i-mean)/std
    normalized_data.append(z_score)
```

```python
sns.histplot(df['total_bill'],kde=True)
```

```python
sns.histplot(normalized_data,kde=True)
```

#### Above standardization using sckitlearn library

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler
```

```python
scaler.fit(df[['total_bill']]) #It will compute mean and standard deviaion
```

```python
scaler.transform(df[['total_bill']])
```

```python
import pandas as pd
pd.DataFrame(scaler.fit_transform(df[['total_bill','tip']]),columns=['Total_bill','Tip'])
```

##### example lets assume that we train our model. and we got some other data we need to transform it in standardization.

```python
import warnings
warnings.simplefilter('ignore')
scaler.transform([[13,4]])
```

#### 2. Normalization- min max scalar

```python
df = sns.load_dataset('taxis')
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()
```

```python
min_max.fit_transform(df[['distance','fare','tip']]) # value between 0 to 1
```

```python
min_max.fit_transform([[1.6,7.0,2.15]])
```
