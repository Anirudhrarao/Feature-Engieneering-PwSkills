### SMOTE (Synthetic minority oversampling technique)

```python
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_samples=1000,n_features=2,n_clusters_per_class=1,weights=[0.90],n_redundant=0,random_state=12)
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(x,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
```

```python
df_final = pd.concat([df1,df2],axis=1)
```

```python
df_final.head()
```

```python
df_final['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df_final['f1'],df_final['f2'],c=df_final['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

#### Transform the dataset

```python
oversample = SMOTE()
```

```python
x,y = oversample.fit_resample(df_final[['f1','f2']],df_final['target'])
```

```python
x.shape,y.shape
```

```python
len(y[y==0])
```

```python
len(y[y==1])
```

```python
df1 = pd.DataFrame(x,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
oversample_df = pd.concat([df1,df2],axis=1)
```
